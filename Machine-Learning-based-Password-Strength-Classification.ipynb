# Checking Password Strength Using ML algorithms

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("data.csv", error_bad_lines = False)
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace = True)
```

```python
passwords_tuple = np.array(data)
```

```python
passwords_tuple
```

```python
import random
random.shuffle(passwords_tuple) # shuffling randomly for robustness
```

```python
y = [labels[1] for labels in passwords_tuple]
```

```python
y
```

```python
X = [labels[0] for labels in passwords_tuple]
```

```python
import seaborn as sns
```

```python
sns.set_style("whitegrid")
sns.countplot(x = 'strength', data = data, palette = 'RdBu_r')
```

```python
def word_divide_char(inputs):
    characters = []
    for i in inputs:
        characters.append(i)
    return characters
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(tokenizer = word_divide_char)
X = vectorizer.fit_transform(X)
```

```python
vectorizer.vocabulary_
```

```python
X
```

```python
data.iloc[0,0]
```

```python
feature_names = vectorizer.get_feature_names()

# get tfidf vector for first document
first_document_vector = X[0]

# print the scores
df = pd.DataFrame(first_document_vector.T.todense(), index = feature_names, columns = ["tfidf"])
df.sort_values(by = ['tfidf'], ascending = False)
```

```python
## logistic regression

from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42) # splitting the data
```

```python
log_class = LogisticRegression(penalty = 'l2', multi_class = 'ovr')
log_class.fit(X_train, y_train)
```

```python
print(log_class.score(X_test, y_test))
```

```python
# Multinomial

clf = LogisticRegression(random_state = 0, multi_class = 'multinomial', solver = 'newton-cg')
clf.fit(X_train, y_train)
print(clf.score(X_test, y_test))
```

```python
X_predict = np.array(["Krish123"])
X_predict = vectorizer.transform(X_predict)
y_pred = log_class.predict(X_predict)
print(y_pred)
```

```python
X_predict = np.array(["Krish"])
X_predict = vectorizer.transform(X_predict)
y_pred = log_class.predict(X_predict)
print(y_pred)
```
